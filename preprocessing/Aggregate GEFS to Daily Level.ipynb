# Aggregate GEFS to Daily Level

Aggregate the GEFS data over the ensemble models, and produce data at the daily level, the same cadence as the Mesonet data.

```python
import pandas as pd
```

Process the train data into daily format

```python
gefs_training_df = pd.read_csv("../data/gefs/gefs_train.csv")
gefs_training_df.head()
```

```python
gefs_training_gb_time_df = (
    gefs_training_df
    .drop(["intTime","intValidTime"], axis=1)
    .groupby(["lat","lon","time"])
    .agg({"mean","min","max","std"})
    .reset_index()
)
gefs_training_gb_time_df.columns = [' '.join(col).strip() for col in gefs_training_gb_time_df.columns.values]
gefs_training_gb_time_df.head()
```

```python
gefs_training_gb_time_df.shape
```

```python
gefs_training_gb_time_df.to_csv("../data/gefs/gefs_train_daily.csv", index=False)
```

Process the test data into daily format

```python
gefs_test_df = pd.read_csv("../data/gefs/gefs_test.csv")
gefs_test_df.head()
```

```python
gefs_test_gb_time_df = (
    gefs_test_df
    .drop(["intTime","intValidTime"], axis=1)
    .groupby(["lat","lon","time"])
    .agg({"mean","min","max","std"})
    .reset_index()
)
gefs_test_gb_time_df.columns = [' '.join(col).strip() for col in gefs_test_gb_time_df.columns.values]
gefs_test_gb_time_df.head()
```

```python
gefs_test_gb_time_df.shape
```

```python
gefs_test_gb_time_df.to_csv("../data/gefs/gefs_test_daily.csv", index=False)
```
